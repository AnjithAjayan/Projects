```python
## Generalize the categorical data i.e. category_code, status and category_code

top10 = list(data.category_code.value_counts().head(10).index)
```

```python
for i in list(data.index):
    if data.category_code[i] not in top10:
        data.category_code.replace(to_replace=data.category_code[i],value='other', inplace=True)
```

```python
dum_cat = pd.DataFrame(pd.get_dummies(data['category_code']))
```

```python
data = pd.concat([data, dum_cat], axis = 1)
data.drop('category_code', axis = 1, inplace=True)
```

```python
top10 = list(data.country_code.value_counts().head(10).index)
```

```python
for i in list(data.index):
    if data.country_code[i] not in top10:
        data.country_code.replace(to_replace=data.country_code[i],value='other', inplace=True)
```

```python
dum_cnty = pd.DataFrame(pd.get_dummies(data['country_code']))
```

```python
data = pd.concat([data, dum_cnty], axis = 1)
data.drop('country_code', axis = 1, inplace=True)
```

```python
for i in list(data.index):
    data.loc[data['status'] == 'operating', 'status'] = 0
    data.loc[data['status'] == 'ipo', 'status'] = 1
    data.loc[data['status'] == 'acquired', 'status'] = 2
    data.loc[data['status'] == 'closed', 'status'] = 3
```

# Feature Engineering

## Mutual Information

```python
from sklearn.feature_selection import mutual_info_regression
```

```python
x = data.drop(['status'], axis=1)
y = data['status']
```

```python
discrete_features = [pd.api.types.is_integer_dtype(t) for t in x.dtypes]
mi_scores = mutual_info_regression(x, y, discrete_features=discrete_features, random_state=0)
mi_scores = pd.Series(mi_scores, name="MI Scores", index=x.columns)
mi_scores = mi_scores.sort_values(ascending=False)
```

```python
plt.figure(dpi=100, figsize=(12, 10))
mi_scores = mi_scores.sort_values()
width = np.arange(len(mi_scores))
ticks = list(mi_scores.index)
plt.barh(width, mi_scores)
plt.yticks(width, ticks)
plt.title("Mutual Information Scores")
plt.show()
```
